## Task 1: Build a unsupervised algorithm to cluster customer data

```python
data=data[data['Income'] < data['Income'].quantile(.99)]
```

```python
data=data[data['Year_Birth'] > data['Year_Birth'].quantile(.01)]
```

```python
data_enc = pd.get_dummies(data, drop_first=True)
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(n_clusters= 2)
```

```python
cluster.fit(data_enc)
```

```python
data_enc['cluster_labels'] = cluster.labels_
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot('Income', 'Year_Birth', data = data_enc, hue = 'cluster_labels', palette='winter');
```

- Clustering the customers using Kmeans
- The clusters of customers are created between the Income of the customers and Year of birth.
- Here we can notice high noise hence, by treating the outliers we can get a clear idea

# Task2: Build a classification model

```python
X = data.drop(['Response'], axis = 1)
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
x = data.drop(['Response'], axis = 1).values
```

```python
y = data[['Response']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

### Logistic Regression

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, xticklabels=['Accepted', 'Not Accepted'], 
                yticklabels=['Accepted', 'Not Accepted']);
```

### XgBoost

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(x_train, y_train)
```

```python
y_pred_final = model.predict(x_test)
predictions = [round(value) for value in y_pred_final]
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

### Decision Tree Classifier

## Running SHAP analysis on the model

- Shap values show how much a given feature changed our prediction (compared to if we made that prediction at some baseline value of that feature).

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# lets get new train and test sets
y = data['Response']
feature_names = [i for i in data.columns if data[i].dtype in [np.int64, np.int64]]
X = data[feature_names]
X = data.drop(['Response'], axis = 1)
```

```python
train_newX, test_newX, train_newy, test_newy = train_test_split(X, y, test_size=0.20, random_state=42)
rf = RandomForestClassifier(random_state=0).fit(train_newX, train_newy)
```

```python
import shap 
explnr = shap.TreeExplainer(rf) # Create object that can calculate shap values
shap_values = explnr.shap_values(test_newX) # now we will calculate the shap values for all the features
shap.summary_plot(shap_values[1], test_newX)
```

- shap value generally gives
    - Vertical location shows what feature it is depicting
    - Color shows whether that feature was high or low for that row of the dataset
    - Horizontal location shows whether the effect of that value caused a higher or lower prediction.

### Based on the SHAP values we can see that NumCatalogPurchases has high recommendated for the model 
- Hence by focusing on the top 3 to 5 shap values we can increase the accuracy of the model
- And by doing so we can get better prediction model
